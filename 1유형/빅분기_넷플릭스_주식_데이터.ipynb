# 넷플릭스 주식 데이터


```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nflx/NFLX.csv")
df.head()
```

Q. 매년 5월달의 open가격의 평균값을 데이터 프레임으로 표현하라



```python
c1 = df[df['Date'].str.contains('-05-')]
c1
```

```python
c1['Date'] = pd.to_datetime(c1['Date'])
```

```python
my_answer = c1.groupby(c1['Date'].dt.strftime('%Y-%m')).mean()['Open'] # strftime > 복습 필요!
my_answer
```

```python
df['Date']  =pd.to_datetime(df['Date'])

target = df.groupby(df['Date'].dt.strftime('%Y-%m')).mean()
answer = target.loc[target.index.str.contains('-05')].Open
print(answer)
```
