# 핸드폰 가격 예측데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv")
df.head()
```

Q. price_range 의 각 value를 그룹핑하여 각 그룹의 n_cores 의 빈도가 가장높은 value와 그 빈도수를 구하여라

```python
my_answer = df[['price_range','n_cores']].groupby(['price_range','n_cores']).size().sort_values().groupby('price_range').tail(1)
my_answer
```

```python
answer =df[['price_range','n_cores']].groupby(['price_range','n_cores']).size().sort_values(0).groupby(level=0).tail(1)
# level=0 > 제일 왼쪽 컬럼을 중심
answer
```

Q. price_range 값이 3인 그룹에서 상관관계가 2번째로 높은 두 컬럼과 그 상관계수를 구하여라

```python
new = df[df['price_range'] == 3].corr().unstack().to_frame().reset_index().dropna().drop_duplicates()
```

```python
new = new.drop(new[new[0] == 1].index)
new
```

```python
new = new.sort_values(by=0, ascending=False).reset_index(drop=True)
new
```

```python
my_answer = new.loc[1,:]
my_answer
```

```python
cordf = df.loc[df['price_range']==3].corr().unstack().sort_values(ascending=False)
answer  = cordf.loc[cordf!=1].reset_index().iloc[1]
print(answer)
```
