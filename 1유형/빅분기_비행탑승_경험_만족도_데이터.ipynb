# 비행탑승 경험 만족도 데이터


```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")
df.head()
```

Q. Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 ‘neutral or dissatisfied’ 보다 ‘satisfied’의 수가 더 높은 Class는 어디 인가?

```python
c1 = df[df['Arrival Delay in Minutes'].isnull()]
c1
```

```python
c2 = c1[['Class','satisfaction']].groupby(['Class','satisfaction']).size().reset_index()
c2
```

```python
c3 = pd.pivot(c2, index='Class', columns='satisfaction')
c3
```

```python
my_answer = c3[c3[0]['neutral or dissatisfied'] < c3[0]['satisfied']]
my_answer
```

```python
answer =df.loc[df['Arrival Delay in Minutes'].isnull()].groupby(['Class','satisfaction'],as_index=False).size().pivot(index='Class',columns='satisfaction')
result =answer[answer['size']['neutral or dissatisfied'] < answer['size']['satisfied']]
result
```
