# 현대 차량 가격 분류문제 데이터


```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/train.csv")
df.head()
```

Q. 정보(row수)가 가장 많은 상위 3차종의 price값의 각 평균값은?



```python
my_answer = df[df['model'].isin(df.value_counts('model').index[:3])].groupby('model').mean()['price'].to_frame()
my_answer
```

```python
answer =df.loc[df.model.isin(df.model.value_counts().index[:3])].groupby('model').mean()['price'].to_frame()
display(answer)
```
