Q. 각년도의 부패에 대한인식을 내림차순 정렬했을때 상위 20개 국가의 부패에 대한인식의 평균을 구하여라

```python
# 1. 모두 기록이 있기 때문에 나라명은 중복될 것이다. 중복되는 행들을 추출한다.
```

```python
new = df[df['나라명'].isin(df[df.duplicated('나라명')]['나라명'].values)]
new
```

```python
# 2. 추출한 것을 2019년으로 한정하고 정리한다.
```

```python
new_2019 = new[new['년도']==2019].sort_values('나라명').reset_index(drop=True)
new_2019
```

```python
# 3. 추출한 것을 2018년으로 한정하고 정리한다.
```

```python
new_2018 = new[new['년도']==2018].sort_values('나라명').reset_index(drop=True)
new_2018
```

```python
# 4. 나라명과 점수 컬럼만 복사해오고 점수 칼럼은 2019년 점수에서 2018년 점수를 뺀 것으로 대체한다.
```

```python
result = new_2019[['나라명','점수']].copy()
result['점수'] = new_2019['점수'] - new_2018['점수']
result
```

```python
# 5. 내림차순 정렬 후 최상위 행을 구한다.
```

```python
result.sort_values('점수',ascending=False).head(1)
```

```python
count = df.나라명.value_counts()
target = count[count>=2].index

df2 =df.copy()
multiple = df2[df2.나라명.isin(target)].reset_index(drop=True)
multiple.loc[multiple['년도']==2018,'점수'] = multiple[multiple.년도 ==2018]['점수'].values * (-1) # 여기서부터 이해 필요!
result = multiple.groupby('나라명').sum()['점수'].sort_values().to_frame().iloc[-1]
result
```
