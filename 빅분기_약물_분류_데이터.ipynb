# 약물 분류 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/train.csv")
df.head()
```

Q. 남성들의 연령대별 (10살씩 구분 0~9세 10~19세 …) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라

```python
c1 = df[df['Sex'] == 'M']
c1
```

```python
my_answer = c1.groupby(df['Age'] // 10 * 10).mean()['Na_to_K'].to_frame()
my_answer
```

```python
pre = df.loc[df.Sex=='M']
pre2= pre.copy()
pre2['Age2'] = pre.Age//10 *10

answer =pre2.groupby('Age2').mean()['Na_to_K'].to_frame()
display(answer)
```
