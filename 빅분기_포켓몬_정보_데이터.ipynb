# 포켓몬 정보 데이터

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/pok/Pokemon.csv')
df.head()
```

Q. Legendary 컬럼은 전설포켓몬 유무를 나타낸다.전설포켓몬과 그렇지 않은 포켓몬들의 HP평균의 차이를 구하여라

```python
my_answer = df[df['Legendary']==True]['HP'].mean() - df[df['Legendary']==False]['HP'].mean()
my_answer
```

```python
target = df.groupby('Legendary').mean()['HP']
result = target.values[1] -target.values[0]
print(result)
```

Q. 가장 많은 Type 1 의 종의 평균 Attack 을 평균 Defense로 나눈값은?

Q. 한번씩만 존재하는 (Type1 , Type2)의 쌍을 각 세대(Generation)은 각각 몇개씩 가지고 있는가?



```python
# 1. 1번만 카운트된 Type 1과 Type 2의 값을 구한다.
```

```python
index = df.value_counts(['Type 1','Type 2'])[df.value_counts(['Type 1','Type 2'])==1].index
index
```

```python
# 2. 각 행의 Type1, Type2 값을 튜플로 묶어주고 인덱스 번호도 가져오기 위해 enumerate 사용한다.
# index에 해당 Type1, Type2 값이 들어있다면 num 리스트에 인덱스 번호를 추가한다.
```

```python
num = []
for i,j in enumerate(zip(df['Type 1'], df['Type 2'])):
  if j in index:
    num.append(i)
```

```python
print(num)
```

```python
# 3. num을 통해 행을 추출하고 Generation 기준으로 카운트를 해준다.
```

```python
my_answer = df.loc[num,:].value_counts('Generation').sort_index()
my_answer
```

```python
target = df[['Type 1','Type 2']].value_counts()
target2 =target[target==1]

lst = []
for value in target2.reset_index().values:
    t1 = value[0]
    t2 = value[1]
    
    sp = df[(df['Type 1']==t1) & (df['Type 2']==t2)]
    lst.append(sp)

result = pd.concat(lst).reset_index(drop=True).Generation.value_counts().sort_index()
print(result)
```
