```python
# Q. 연령대 코드가 20대인 데이터를 추출하고,이동거리값이 추출한 데이터의 이동거리값의 평균 이상인 데이터를 추출한다.
# 최종 추출된 데이터를 대여일자, 대여소 번호 순서로 내림차순 정렬 후 1행부터 200행까지의 탄소량의 평균을 소숫점 3째 자리까지 구하여라
```

```python
# 1. 연령대 코드 20대 추출
```

```python
c1 = df[df['연령대코드'] == '20대']
c1
```

```python
# 2. 이동거리값이 추출한 데이터의 이동거리값의 평균 이상인 데이터를 추출
```

```python
c2 = c1[c1['이동거리'] >= c1['이동거리'].mean()]
c2
```

```python
# 3. 최종 추출된 데이터를 대여일자, 대여소 번호 순서로 내림차순 정렬
```

```python
c3 = c2.sort_values(['대여일자','대여소번호'], ascending=False)
c3
```

```python
# 4. 평균을 구하려고 하였으나 탄소량의 타입이 object라서 오류 발생 > float으로 변경
```

```python
c3.info()
```

```python
c3['탄소량'] = c3['탄소량'].astype(float)
```

```python
# 5. 1행부터 200행까지의 탄소량의 평균을 소숫점 3째 자리까지 구하여라
```

```python
my_answer = c3[:200]['탄소량'].mean().round(3)
my_answer
```

```python
tw = df[df.연령대코드 =='20대'].reset_index(drop=True)
tw_mean = tw[tw.이동거리 >= tw.이동거리.mean()].reset_index(drop=True)
tw_mean['탄소량'] =tw_mean['탄소량'].astype('float')
target =tw_mean.sort_values(['대여일자','대여소번호'], ascending=False).reset_index(drop=True).iloc[:200].탄소량
result = round(target.sum()/len(target),3)
print(result)
```

```python
c1.groupby(['대여시간','대여소번호']).count().sort_values('이용건수',ascending=False)
```

```python
target = df[(df.day_name.isin(['Tuesday', 'Wednesday', 'Thursday', 'Friday','Monday'])) & (df.대여시간.isin([6,7,8]))]
result = target.groupby(['대여시간','대여소번호']).size().to_frame('이용 횟수') # to_frame('컬럼명')
answer = result.sort_values(['대여시간','이용 횟수'],ascending=False).groupby('대여시간').head(3) # groupby와 head
display(answer)
```

```python
# Q. 남성(‘M’ or ‘m’)과 여성(‘F’ or ‘f’)의 이동거리값의 평균값을 구하여라
```

```python
df['성별'].unique()
```

```python
df['성별'] = df['성별'].str.replace('M','남성')
df['성별'] = df['성별'].str.replace('m','남성')
df['성별'] = df['성별'].str.replace('F','여성')
df['성별'] = df['성별'].str.replace('f','여성')
```

```python
df[['성별','이동거리']].groupby('성별').mean()
```

```python
df['sex'] = df['성별'].map(lambda x: '남' if x in ['M','m'] else '여') # 추가 공부 필요!
answer = df[['sex','이동거리']].groupby('sex').mean()
display(answer)
```
